#### IMPORT INITIAL DATA TO DATAFRAMES

```python
address_file = pd.read_csv("panama_papers.nodes.address.csv")
entity_file = pd.read_csv("panama_papers.nodes.entity.csv")
intermediary_file = pd.read_csv("panama_papers.nodes.intermediary.csv")
officer_file = pd.read_csv("panama_papers.nodes.officer.csv")
edges_file = pd.read_csv("panama_papers.edges.csv")
```

#### CREATING THE SQL TABLE *"dit2122_countries"*
